```python
import numpy as np
import matplotlib.pyplot as plt
data = np.arange(1,50)
noise = np.random.randint(-10, 10, len(data))
noisy_data = data + noise
```

# Making data and adding noise to signal

```python
fig, ax = plt.subplots(1,2)
ax[0].plot(data)
ax[1].plot(noisy_data)
```

```python
noisy_data = [[i, noisy_data[i]] for i in range(len(noisy_data))]
```

# Applying PCA

```python
import numpy as np
from sklearn.decomposition import PCA
pca = PCA(n_components=1)
recovered_data = pca.fit_transform(noisy_data)
```

```python
fig, ax = plt.subplots(1,3, figsize=(15, 5))
ax[0].scatter(data,data)
ax[0].title.set_text('data')
ax[1].scatter(np.array(noisy_data)[:,0], np.array(noisy_data)[:,1])
ax[1].title.set_text('noisy data')
ax[2].scatter(data, recovered_data[:,0])
ax[2].title.set_text('projected')
```

```python
print(pca.explained_variance_ratio_)
```

```python
data = np.sin(np.arange(.1,6.28,.1))
noise = np.random.rand(len(data))
noisy_data = data + noise
noisy_data = [[i, noisy_data[i]] for i in range(len(noisy_data))]
pca = PCA(n_components=1)
recovered_data = pca.fit_transform(noisy_data)
```

```python
fig, ax = plt.subplots(1,3, figsize=(15, 5))
ax[0].scatter(np.arange(.1,6.28,.1),data)
ax[1].scatter(np.arange(.1,6.28,.1), np.array(noisy_data)[:,1])
ax[2].scatter(np.arange(.1,6.28,.1), recovered_data[:,0])
```

```python
print(pca.explained_variance_ratio_)
```
